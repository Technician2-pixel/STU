```python
print("======STUDENT PROFILE======")

name_input = input("Enter full name: ")          
age_input = input("Enter age: ")                 
gpa_input = input("Enter GPA: ")              
courses_input = input("Enter courses (comma-separated): ")  

age = int(age_input)            
gpa = float(gpa_input)            
courses = courses_input.split(",")  
courses = [course.strip() for course in courses]

student_profile = {
    "name": name_input,
    "age": age,
    "gpa": gpa,
    "courses": courses,
    "num_courses": len(courses)
}

on_deans_list = gpa >= 3.5
is_minor = age < 18


print("======STUDENT PROFILE======")
print(f"Name:        {student_profile['name']}")
print(f"Age:         {student_profile['age']} ({'Minor' if is_minor else 'Adult'})")
print(f"GPA:         {student_profile['gpa']}")
print(f"Courses:     {student_profile['courses']}")
print(f"# Courses:   {student_profile['num_courses']}")
print(f"Dean's List: {'Yes' if on_deans_list else 'No'}")
print("===========================")


print("======== TYPE DEMONSTRATION======")
print(f" age input type: {type(age_input)}  Converted type: {type(age)}")
print(f"GPA input type: {type(gpa_input)}  Converted type: {type(gpa)}")
print(f" courses input type: {type(courses_input)}  Converted type: {type(courses)}")
```
